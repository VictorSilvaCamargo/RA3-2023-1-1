```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Definindo a função
x = sp.symbols('x')
f = 40*x**3 + 12*x**2 - 9*x

# Intervalo de integração
a = 2.5
b = 7

# Calculando a área usando álgebra
area = sp.integrate(f, (x, a, b))

# Convertendo a função para uma função numpy
f_np = sp.lambdify(x, f)

# Plotando o gráfico da função
x_vals = np.linspace(a, b, 100)
y_vals = f_np(x_vals)

plt.plot(x_vals, y_vals, label='f(x)')
plt.fill_between(x_vals, y_vals, where=((a <= x_vals) & (x_vals <= b)), alpha=0.5)

# Marcando a área entre os limites de integração
plt.axvline(x=a, color='red', linestyle='--', label='Limite Inferior')
plt.axvline(x=b, color='green', linestyle='--', label='Limite Superior')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função')
plt.legend()
plt.show()

# Imprimindo o resultado da área
print("A área sobre a curva é:", area)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols
from ipywidgets import interact, IntSlider

# Define a variável e a função
x = symbols('x')
f = 40*x**3 + 12*x**2 - 9*x

# Converte f para uma função numérica para cálculo e plotagem
funcao = lambdify(x, f, 'numpy')

# Limites de integração
a, b = 2.5, 7

# Calcula a integral definida
integral_value = float(integrate(f, (x, a, b)))

def plot_riemann_sums(N):
    dx = (b-a)/N
    x = np.linspace(a, b, N+1)
    y = funcao(x)

    plt.figure(figsize=(7,5))

    x_right = x[:-1] 
    y_right = y[:-1]
    plt.bar(x_right, y_right, width=dx, alpha=0.2, align='edge', edgecolor='gray')

    # Criar um conjunto separado de valores x e y para a curva
    x_curve = np.linspace(a, b, 1000)
    y_curve = funcao(x_curve)

    # Plot a curva depois dos retângulos para que ela apareça por cima
    plt.plot(x_curve, y_curve, 'blue')

    riemann_sum = np.sum(funcao(x[:-1]) * dx)
    error = np.abs(riemann_sum - integral_value)

    plt.title('Soma de Riemann, N = {}, Erro = {:.4f}'.format(N, error))

    # Traça os eixos x e y
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.show()

    print("Área sobre a curva: ", riemann_sum)

interact(plot_riemann_sums, N=IntSlider(min=10, max=1000, step=10, value=10))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, lambdify, integrate
from ipywidgets import interact, IntSlider

# Define a variável e a função
x = symbols('x')
f = 7*sp.sin(x) - 2*sp.cos(x)

# Converte f para uma função numérica para cálculo e plotagem
funcao = lambdify(x, f, 'numpy')

# Limites de integração
a, b = 0, np.pi/2

# Calcula a integral definida
integral_value = float(integrate(f, (x, a, b)))

def plot_riemann_sums(N):
    dx = (b-a)/N
    x_vals = np.linspace(a, b, N+1)
    y_vals = funcao(x_vals)

    plt.figure(figsize=(7,5))

    x_right = x_vals[:-1] 
    y_right = y_vals[:-1]
    areas = y_right * dx  # Calcula as áreas dos retângulos

    plt.bar(x_right, y_right, width=dx, alpha=0.2, align='edge', edgecolor='gray')

    # Cria um conjunto separado de valores x e y para a curva
    x_curve = np.linspace(a, b, 1000)
    y_curve = funcao(x_curve)

    # Plota a curva depois dos retângulos para que ela apareça por cima
    plt.plot(x_curve, y_curve, 'blue')

    riemann_sum = np.sum(areas)
    error = np.abs(riemann_sum - integral_value)

    plt.title('Soma de Riemann, N = {}, Erro = {:.4f}'.format(N, error))

    # Traça os eixos x e y
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)

    plt.xlim(a, b)  # Define os limites do eixo x
    plt.ylim(-2, plt.ylim()[1])  # Define os limites do eixo y

    plt.show()

    print("Área sob a curva: ", riemann_sum)

interact(plot_riemann_sums, N=IntSlider(min=10, max=1000, step=10, value=10))
```
